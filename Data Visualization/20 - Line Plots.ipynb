```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
df = fmri.copy()
df.head()
```

### Creating Line Plots

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = df)
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             style = "event", 
             hue = "event",
             markers = True,
             dashes = False,
             data = df);
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             style = "event", 
             hue = "region",
             data = df);
```
