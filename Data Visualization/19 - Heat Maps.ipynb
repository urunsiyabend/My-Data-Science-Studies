# Heat Maps

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.info()
```

```python
df["passengers"].describe()
```

```python
df = df.pivot("month", "year", "passengers");
```

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df, cmap = "mako");
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = .5);
```
