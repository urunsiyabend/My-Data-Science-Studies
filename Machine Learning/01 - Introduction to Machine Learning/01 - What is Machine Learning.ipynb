# What is Machine Learning

#### Machine Learning is a field of study for developing various algorithms and techniques to enable computers to learn like humans.

## Terminology

**Independent Variable:** The variable which affects other variables.

**Dependent Variables:** The variable which affects from other variables.

**Types of Learning:**
* Supervised Learning
* Semi-supervised Learning
* Unsupervised Learning

**Types of Problems:**
* Regression
* Classification

**Variable Types:**
* Numeric Variables
* Categorical Variables

**Scales of Measurement:**
* **For Numeric Variables:** Interval and Ratio
* **For Categorical Variables:** Nominal and Ordinal

**Test-Train Split**: Splitting original data to training and testing the machine learning model

**Feature Engineering**

**Variable Selection**

**Model Selection**

**Overfitting**

**Types of Models**:
* Deterministic Models
* Stochastic Models



**Linearity**:
* Linear Models
* Nonlinear Models
