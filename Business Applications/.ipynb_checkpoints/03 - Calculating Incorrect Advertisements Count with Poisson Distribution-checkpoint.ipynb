# Calculating Probability of Incorrect Advertisement Entries with Poisson Distribution 

**Problem:**
* It is desired to calculate the probability of incorrect advertisement entries.

**Details:**
* Measurements were taken for one year.
* Distribution is Poisson and lambda value is 0,1 (Average error count)
* Probability of no incorrect entry, 3 incorrect entries and 5 incorrect entries?

```python
from scipy.stats import poisson
```

```python
lmb = 0.1
```

```python
rv = poisson(mu = lmb)
print(rv.pmf(k = 0))
```

```python
print(rv.pmf(k = 3))
```

```python
print(rv.pmf(k = 5))
```
