# Advertisement Expense Optimization with Binomial Distribution

### Problem:
* Advertisements are placed on various media.
* Working on optimizations about ad-clicks and feed-back rates.
* It is desired to calculate the probability of ad-clicks for various scenarios.

### Details:
* An advertisement will be placed.
* Distribution and ad-click probabilities are known. (0,01)
* What is the ad-click probability when ad was viewed by 100 people. (1,5,10)

```python
from scipy.stats import binom
```

```python
p = 0.01
n = 100
rv = binom(n, p)
print(rv.pmf(1))
print(rv.pmf(5))
print(rv.pmf(10))
```
