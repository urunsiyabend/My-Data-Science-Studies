# Conversion Rate Test with Single Sample Proportion Test

**Problem:**
* It is placed an advertisement with an agent and it is expressed the conversion rate is 0.125 by advertisement agent. But this value doesn't fit with incomes and it is desired to test the conversion rate. 

**Details**:
* Our advertisemnt is clicked by 500 person and 40 of them did shopping in our website.
* The conversion rate obtained from the sample: 40 / 500 = 0.08

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = 40
nobs = 500
value = 0.125
```

```python
proportions_ztest(count, nobs, value)
```

## Explanation:

#### Achieved pvalue (0.002) is lower than alpha (0.05) so we can reject hypothesis H0.
