# User Interface Experiment with AB Test

**Problem:**
* There are two website designs with different colorized buttons.
* It is desired to choose the more attractive design to improve click rate.

**Details:**
* There are measuremnts collected.
* Blue Button: 1000 views, 520 clicks
* Green Button: 680 views, 489 clicks

## Hypotheses

## Application

```python
from statsmodels.stats.proportion import proportions_ztest
import pandas as pd
import numpy as np
```

```python
success_counts = np.array([520, 489])
obs_counts = np.array([1000, 489])
```

```python
proportions_ztest(count = success_counts, nobs = obs_counts)
```
