# Session Duration Test with One Sample T Test

**Problem**:
* Is it 170 seconds the average session duration, indeed?

**Details**:
* There are average durations that collected with softwares.

```python
import numpy as np
measurements = np.random.normal(154, 50, size = (1, 50))[0]
measurements
```

```python
measurements.std()
```

```python
measurements.mean()
```

## Assumption Control

```python
import scipy.stats as stats
import pandas as pd
```

```python
stats.describe(measurements)
```

```python
# Assumptions
# Normality Assumption
```

```python
pd.DataFrame(measurements).plot.hist()
```

```python
import pylab
stats.probplot(measurements, dist = "norm", plot = pylab)
pylab.show()
```

## Shapiro-Wilks Test

**H0: There is no significant difference between sampling distribution and theoritical normal distribution.**

**H1: There is significant difference between sampling distribution and theoritical normal distribution.**

```python
from scipy.stats import shapiro
```

```python
shapiro(measurements)
```

#### Calculated p-value is not lower than alpha (0.05), so we can not reject hypothesis H0

## Applying Hypothesis Test

```python
stats.ttest_1samp(measurements, popmean= 170)
```

**H0: Average session duration is 170 seconds**

**H1: Average session duration is not 170 seconds**

## Explanation

#### Calculated p-value is lower than alpha (0.05) so we can reject hypothesis H0
