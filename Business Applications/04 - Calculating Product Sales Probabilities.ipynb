# Calculating Product Sales Probabilities with Normal Distribution

**Problem:**
* It is desired to calculate the probabilities of next month sales.


**Details:**
* Monthly average sales = 80K, standart deviation = 5K
* What is the probability of making sales over 90K?

```python
from scipy.stats import norm
```

### Over 90K

```python
1-norm.cdf(90, 80, 5)
```

### Over 70K

```python
1-norm.cdf(70, 80, 5)
```

### Under 73K

```python
norm.cdf(73, 80, 5)
```

### Between 85 and 90

```python
norm.cdf(90, 80, 5) - norm.cdf(85, 80, 5)
```
