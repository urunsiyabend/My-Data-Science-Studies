# AdvancedAggregate Functions

* aggregate
* filter
* transform
* apply

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
        {
            'groups': ['a', 'b', 'c', 'a', 'b', 'c'],
            'variable1': np.random.rand(1, 6)[0] * 100,
            'variable2': np.random.rand(1, 6)[0] * 100,
        }
)
```

```python
df
```

```python
df.groupby('groups').describe().T
```

```python
df.groupby('groups').aggregate(['%50', np.median, 'max']).T
```
