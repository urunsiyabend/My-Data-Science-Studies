# Variable Transformation

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

## 0-1 Tranformation (Label Encoding)

```python
from sklearn import preprocessing
```

```python
lbe = preprocessing.LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["sex_new"] = lbe.fit_transform(df["sex"])
df
```

## 1 and Others Transformation

```python
import numpy as np
df["day_new"] = np.where(df["day"].str.contains("Sun"), 1, 0)
df
```

## Multiple Class Transformation

```python
from sklearn import preprocessing
```

```python
lbe = preprocessing.LabelEncoder()
```

```python
lbe.fit_transform(df["day"])
```
