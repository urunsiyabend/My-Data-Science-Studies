# Grouping

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'groups': ['a', 'b', 'c', 'a', 'b', 'c'],
    'data': [345, 34, 23, 45, 243, 23],
},columns=['groups', 'data'])
```

```python
df
```

```python
df.groupby("groups")
```

```python
df.groupby('groups').sum()
```
